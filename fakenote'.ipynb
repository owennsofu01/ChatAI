```python
# ===============================================================
# 1) INSTALL + IMPORT LIBRARIES
# ===============================================================
!pip install kagglehub --quiet

import kagglehub
import os
import numpy as np
import cv2
import matplotlib.pyplot as plt
from sklearn.metrics import confusion_matrix, classification_report

import tensorflow as tf
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten, Dense, Dropout
from tensorflow.keras.optimizers import Adam

# ===============================================================
# 2) DOWNLOAD DATASET FROM KAGGLE
# ===============================================================
path = kagglehub.dataset_download("gauravsahani/indian-currency-notes-classifier")
print("Dataset downloaded to:", path)

train_path = os.path.join(path, "Train")
valid_path = os.path.join(path, "Test")

IMAGE_SIZE = (224, 224)

# ===============================================================
# 3) DATA GENERATOR (Normalization + Augmentation)
# ===============================================================
datagen = ImageDataGenerator(
    rescale=1./255,
    shear_range=0.15,
    zoom_range=0.15,
    horizontal_flip=True
)

train_set = datagen.flow_from_directory(
    train_path,
    target_size=IMAGE_SIZE,
    batch_size=32,
    class_mode='categorical'
)

val_set = datagen.flow_from_directory(
    valid_path,
    target_size=IMAGE_SIZE,
    batch_size=32,
    class_mode='categorical'
)

# ===============================================================
# 4) ADVANCED IMAGE PREPROCESSING FUNCTION (OpenCV)
# ===============================================================
def preprocess_image(path):
    img = cv2.imread(path)
    img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    img = cv2.GaussianBlur(img, (3, 3), 0)
    img = cv2.convertScaleAbs(img, alpha=1.3, beta=15)
    kernel = np.array([[-1,-1,-1], [-1,9,-1], [-1,-1,-1]])
    img = cv2.filter2D(img, -1, kernel)
    img = cv2.resize(img, (224,224))
    img = img.astype("float32") / 255.0
    img = np.expand_dims(img, axis=0)
    return img

# ===============================================================
# 5) BUILD CNN MODEL
# ===============================================================
model = Sequential([
    Conv2D(32, (3,3), activation="relu", input_shape=(224,224,3)),
    MaxPooling2D((2,2)),

    Conv2D(64, (3,3), activation="relu"),
    MaxPooling2D((2,2)),

    Conv2D(128, (3,3), activation="relu"),
    MaxPooling2D((2,2)),

    Flatten(),
    Dense(128, activation="relu"),
    Dropout(0.3),
    Dense(train_set.num_classes, activation="softmax")
])

model.compile(
    optimizer=Adam(learning_rate=0.0005),
    loss="categorical_crossentropy",
    metrics=["accuracy"]
)

model.summary()

# ===============================================================
# 6) TRAIN MODEL
# ===============================================================
history = model.fit(
    train_set,
    validation_data=val_set,
    epochs=10
)

model.save("currency_classifier_cnn.h5")
print("CNN model saved!")

# ===============================================================
# 7) PREDICT FUNCTION (USING ADVANCED PREPROCESSING)
# ===============================================================
def predict_image(path):
    img = preprocess_image(path)
    prediction = model.predict(img)
    index = np.argmax(prediction)
    labels = list(train_set.class_indices.keys())
    confidence = prediction[0][index]
    print(f"Prediction: {labels[index]} ({confidence*100:.2f}% confidence)")
    return labels[index]

# Example usage:
# predict_image("/content/test.jpg")

# ===============================================================
# 8) CONFUSION MATRIX + CLASSIFICATION REPORT
# ===============================================================
y_true = val_set.classes
y_pred = []

val_set.reset()
for i in range(len(val_set)):
    preds = model.predict(val_set[i][0])
    y_pred.extend(np.argmax(preds, axis=1))

labels = list(train_set.class_indices.keys())

cm = confusion_matrix(y_true, y_pred)
print("\nConfusion Matrix:\n", cm)

print("\nClassification Report:\n")
print(classification_report(y_true, y_pred, target_names=labels))

plt.figure(figsize=(6,5))
plt.imshow(cm, cmap="Blues")
plt.title("Confusion Matrix")
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.xticks(np.arange(len(labels)), labels, rotation=45)
plt.yticks(np.arange(len(labels)), labels)
plt.colorbar()
plt.show()

# ===============================================================
# 9) EXPORT CNN MODEL TO TFLITE
# ===============================================================
converter = tf.lite.TFLiteConverter.from_keras_model(model)
tflite_model = converter.convert()

with open("currency_classifier_cnn.tflite", "wb") as f:
    f.write(tflite_model)

print("TFLite CNN model saved!")
```
